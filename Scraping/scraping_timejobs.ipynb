```python
from bs4 import BeautifulSoup
import requests
```

```python
link = requests.get('https://m.timesjobs.com/jobskill/%2522java+developer%2522%252c%2522php+developer%2522%252c%2522android+developer%2522%252c%2522content+writer%2522%252c%2522business+development+manager%2522-jobs')
print(link)
```

```python
soup = BeautifulSoup(link.text, 'lxml')
soup
```

```python
jobs = soup.find('li')
jobs
```

```python
jobs.h4.span.text
```

```python
for job in jobs:
    texts = jobs.find('div').text.strip()
    print(texts)
```
